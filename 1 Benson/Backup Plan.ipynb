```python
# strip column names from extra white space
df1.columns = [column.strip() for column in df1.columns]
df1.columns
```

```python
# setting to data series
```

```python
df1['DATE_TIME'] = pd.to_datetime(df1.DATE + df1.TIME, format="%m/%d/%Y%H:%M:%S")
```

```python
df1.head()
```

```python
# check for duplicates
(df1
 .groupby(["C/A", "UNIT", "SCP", "STATION", "DATE_TIME"])
 .ENTRIES.count()
 .reset_index()
 .sort_values("ENTRIES", ascending=False)).head(15)
```

```python
# we need to drop these.
df1.sort_values(['C/A', 'UNIT', 'SCP', 'STATION', 'DATE_TIME'], inplace=True, ascending=False)
```

```python
df1.drop_duplicates(subset=['C/A', 'UNIT', 'SCP', 'STATION', 'DATE_TIME'], inplace=True)
```

```python
# let's check those duplicates again
(df1
 .groupby(["C/A", "UNIT", "SCP", "STATION", "DATE_TIME"])
 .ENTRIES.count()
 .reset_index()
 .sort_values("ENTRIES", ascending=False)).head(5)
```

```python
# let's see if there are inconsistencies in the way lines are recorded
df1.LINENAME.unique()
```

```python
len(df1.LINENAME.unique())
```

```python
# looks like there are duplicates like NQR456W and 456NQRW
# we need this to merge with our station localization data
# we can sort it to fix it
df1.LINENAME = df1.LINENAME.apply(lambda x : ''.join(sorted(x)))
```

```python
len(df1.LINENAME.unique())
```

```python
# looks like we took take care of the ordering for about 20 obs.
```

```python
# we can now drop some columns we don't need
df1 = df1.drop(['DIVISION', 'DESC'], axis=1, errors='ignore')
df1.head()
```

```python
# sort by DATE_TIME and group by
df1 = df1.sort_values('DATE_TIME')
```

```python
# create offsets for ENTRIES EXITS and DATE_TIME
df1[['PREV_DATE_TIME', 'PREV_ENTRIES', 'PREV_EXITS']] = \
    (df1.groupby(['C/A', 'UNIT', 'SCP', 'STATION'])
        ['DATE_TIME','ENTRIES', 'EXITS'].shift())
```

```python
df1
```

```python
# any counters going in the wrong direction
len(df1[df1['ENTRIES'] < df1['PREV_ENTRIES']])
```

```python
# is this a significant portion of the data?
len(df1[df1['ENTRIES'] < df1['PREV_ENTRIES']]) / len(df1)
```

```python
# we drop NaNs and NaTs
df1.dropna(subset=['PREV_DATE_TIME'], inplace=True)
df1.head(5)
```

```python
df2 = df1[df1.DATE_TIME - df1.PREV_DATE_TIME == timedelta(hours=4)]
```

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 25)
pd.set_option('display.max_rows', 100)
pd.set_option('display.precision', 3)
```

```python
# read in the file from nyc open data
df = pd.read_csv('NYC_Transit_Subway_Entrance_And_Exit_Data.csv', dtype=str)
df.head()
```

```python
#remove line names that are empty to prepare for concatenation
df = df.fillna('')
df
```

```python
df.Route11.unique()
```

```python
# Concantenate and sort to create new column Line Name
df['Line Name'] = df.Route1 + df.Route2 + df.Route3 + df.Route4 \
                  + df.Route5 + df.Route6 + df.Route7 + df.Route8 \
                  + df.Route9 + df.Route10 + df.Route11
df['Line Name'] = df['Line Name'].apply(lambda x : ''.join(sorted(x)))
```

```python
# only keep desired columns
df = df[['Station Name', 'Line Name', 'Station Longitude', 'Station Latitude']]
df
```

```python
df['Line Name'].value_counts()
```

```python
df['Line Name'] = df['Line Name'].apply(lambda x : x.strip())
```

```python
df['Line Name'].value_counts()
```

```python
# drop duplicates left over from multiple entrances
df = df.drop_duplicates()
df = df.reset_index()
len(df)
```

```python
# drop the newly created index 
df = df.drop('index', axis=1)
df
```

```python
# write dataframe to csv
df.to_csv('stations_location', encoding='utf-8', index=False)
```

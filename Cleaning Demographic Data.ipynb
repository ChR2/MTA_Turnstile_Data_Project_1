```python
# https://data.cityofnewyork.us/api/views/kku6-nxdu/rows.csv?accessType=DOWNLOAD
# https://www.irs.gov/statistics/soi-tax-stats-individual-income-tax-statistics-2015-zip-code-data-soi
import pandas as pd
import numpy as np
import uszipcode
import csv
from collections import defaultdict
import folium
```

```python


pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', 250)
pd.set_option('display.precision', 3)

```

```python
#input: csv with zip codes nyc
# out: list of nyc zip codes in string format
with open('All_NYC_ZipCodes.csv', 'r') as f: #from USPS post office
    reader = csv.reader(f)
    next(reader)
    nyc_zip = list(reader)

nyc_zip = [i[0] for i in nyc_zip]
print(nyc_zip)
```

```python
# df['Number of returns'].value_counts()
# df.iloc[:,2].value_counts()
```

```python
# call zipcode search engine
from uszipcode import ZipcodeSearchEngine
search = ZipcodeSearchEngine()

```

```python
zipcode = search.by_zipcode('10002')
print(zipcode)
```

```python
# input: zipcodes
# output: dictionary with zipcode as key and list of Wealthy, Latitudes as values
my_dict = defaultdict(int)

for i in nyc_zip:
    zipcode = search.by_zipcode(i)
    
    my_dict[i] = [zipcode.Zipcode,zipcode.Wealthy,zipcode.NEBoundLatitude,zipcode.NEBoundLongitude,zipcode.SWBoundLatitude, zipcode.SWBoungLongitude]

print(my_dict)
```

```python
new_list = my_dict.values()

```

```python
new_list #list of lists of zip code info
```

```python
labels = ['Zipcode','Wealthy', 'NEBoundLatitude', 'NEBoundLongitude','SWBoundLatitude', 'SWBoungLongitude']
```

```python
#make list of tuples
wtf= [tuple(i) for i in new_list]
print(wtf)
```

```python
#input: list of tuples
#output: dataframe. finally
df = pd.DataFrame.from_records(wtf, columns=labels)
```
